e. Suppose we knew that the sample came from a Poisson distribution, but the parameter is unknown.

## answer
يقوم هذا الكود بحساب متوسط ​​العينة باستخدام np.mean(data_pois). ثم يحسب الاحتمالات البواسونية للقيم الفريدة في العينة باستخدام الصيغة التي تنطبق على توزيع بواسون. يتم طباعة التكرارات الملاحظة من جدول التكرار وقيم الاحتمالات البواسونية إلى وحدة التحكم للمقارنة.

sample_mean = np.mean(data_pois)
تعمل np.mean() على حساب المتوسط الحسابي لمصفوفة البيانات. يتم إعطاء مصفوفة البيانات data_pois كمدخل لـ np.mean()، وتقوم الدالة بحساب القيمة المتوسطة لهذه المصفوفة.

factorial_func = np.vectorize(np.math.factorial)
دالة غير متجهة (scalar function) هي دالة تستقبل وتعيد قيمة واحدة في كل مرة. عند استدعاء الدالة مع مدخل واحد، فإنها تقوم بتنفيذ العمليات المطلوبة وتعيد قيمة واحدة كنتيجة لهذه العملية. مثال على دالة غير متجهة هو np.math.factorial(n) التي تأخذ عددًا صحيحًا n كمدخل وتعيد عاملي n! كنتيجة واحدة.

بالمقابل، دالة متجهة (vectorized function) هي دالة تستقبل مصفوفة أو سلسلة من القيم كمدخل وتعيد مصفوفة أو سلسلة أخرى كنتيجة. عند استدعاء الدالة مع مصفوفة كمدخل، فإنها تقوم بتنفيذ العمليات المطلوبة على كل عنصر في المصفوفة وتعيد مصفوفة جديدة تحتوي على النتائج المحسوبة لكل عنصر. هذا يسمح بتطبيق العمليات على مجموعة من البيانات في وقت واحد بدلاً من تطبيقها على كل عنصر بشكل منفرد. تسمح دوال NumPy المتجهة مثل np.exp() و np.power() بتطبيق العمليات بشكل متجه على مصفوفة القيم.


في الشيفرة المذكورة poisson_probabilities = np.exp(-sample_mean) * np.power(sample_mean, unique_values) / factorial_func(unique_values)، يتم حساب قيم احتمالات بواسون (Poisson probabilities) باستخدام الصيغة الرياضية لتوزيع بواسون.

تفسير العمليات في هذه الشيفرة:

np.exp(-sample_mean): تستخدم دالة np.exp() لحساب القيمة العددية للمقدار e مرفوع للأس السالب (-sample_mean). يتم استخدام هذا المقدار في الصيغة الرياضية لتوزيع بواسون.

np.power(sample_mean, unique_values): تستخدم دالة np.power() لحساب القيمة العددية لـ sample_mean مرفوع لأس قيم unique_values. هذه العملية تقوم بحساب المقدار (sample_mean)^k في الصيغة الرياضية لتوزيع بواسون، حيث k يمثل القيمة في unique_values.

factorial_func(unique_values): تستخدم الدالة المتجهة factorial_func() المنشئة سابقًا لحساب عاملي القيم في unique_values. تُطبَّق دالة np.math.factorial() على كل عنصر في unique_values باستخدام factorial_func().

بعد ذلك، يتم قسمة الناتج من العمليات الأولى والثانية على نتائج عاملي القيم في unique_values باستخدام factorial_func(unique_values).

النتيجة النهائية ستكون مصفوفة poisson_probabilities تحتوي على احتمالات بواسون المحسوبة لكل قيمة في unique_values.
####
np.column_stack() هي دالة في مكتبة NumPy تستخدم لدمج قوائم من المصفوفات عموديًا لإنشاء مصفوفة جديدة. في السياق الحالي، تستخدم np.column_stack() لدمج مصفوفتي unique_values و poisson_probabilities عموديًا.

في الشيفرة np.column_stack((unique_values, poisson_probabilities))، يتم تمرير قائمة تحتوي على المصفوفتين unique_values و poisson_probabilities إلى np.column_stack(). الدالة تقوم بدمج هاتين المصفوفتين عموديًا لإنشاء مصفوفة جديدة تحتوي على القيم المتجاورة من كلا المصفوفتين.

على سبيل المثال، إذا كانت unique_values تحتوي على [1, 2, 3, 4] و poisson_probabilities تحتوي على [0.25, 0.5, 0.75, 1.0]، فإن الشيفرة np.column_stack((unique_values, poisson_probabilities)) ستنتج المصفوفة التالية:

```python
sample_mean = np.mean(data_pois)

factorial_func = np.vectorize(np.math.factorial)
poisson_probabilities = np.exp(-sample_mean) * np.power(sample_mean, unique_values) / factorial_func(unique_values)

print("Observed Frequencies:\n", frequency_table)
print("\nPoisson Probabilities:\n", np.column_stack((unique_values, poisson_probabilities)))
```
