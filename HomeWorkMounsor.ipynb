b. Using the data for the five experiments together, plot the empirical distribution function for the data.

ب. باستخدام بيانات التجارب الخمس معًا، ارسم دالة التوزيع التجريبي للبيانات.

Calculate the empirical distribution function (EDF):
حساب وظيفة التوزيع العملية (EDF):

Here, edf_x contains the sorted speed values, and edf_y represents the probabilities associated with each value.
هنا، edf_x يحتوي على القيم المرتبة للسرعة، و edf_y يمثل الاحتمالات المرتبطة بكل قيمة.


في هذا الشفرة، يتم استخدام plt.plot لإنشاء الرسم الشرطي لوظيفة التوزيع العملية (EDF). الخيار drawstyle='steps-post' يستخدم لإنشاء رسم شبه درجاتي حيث يقفز الرسم عند كل نقطة بيانات.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('michelson.csv')
speed_data = df['Speed']
n = len(speed_data)
edf_x = np.sort(speed_data)
edf_y = np.arange(1, n + 1) / n
plt.plot(edf_x, edf_y, drawstyle='steps-post')
plt.xlabel('Speed of Light (km/s)')
plt.ylabel('Probability')
plt.title('Empirical Distribution Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)


S_values = []
T_values = []


for _ in range(1000):

    sample = np.random.exponential(scale=1/2, size=20)
    S = 1 / np.mean(sample)
    T = 20 * np.min(sample)
    S_values.append(S)
    T_values.append(T)

S_values = np.array(S_values)
T_values = np.array(T_values)


plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(S_values, bins=30, color='blue', alpha=0.7)
plt.xlabel('S')
plt.ylabel('Frequency')
plt.title('Histogram of S')
plt.subplot(1, 2, 2)
plt.hist(T_values, bins=30, color='red', alpha=0.7)
plt.xlabel('T')
plt.ylabel('Frequency')
plt.title('Histogram of T')
plt.tight_layout()
plt.show()

var_S = np.var(S_values)
var_T = np.var(T_values)

print("Variance of S:", var_S)
print("Variance of T:", var_T)
```

e. Calculate the empirical variances for S and T based on the original sample. Which one has smaller
variance?
ه. حساب الفروق التجريبية لS وT استنادا إلى العينة الأصلية. أيهما أصغر

To calculate the empirical variances for S and T based on the original sample, you can use the numpy library in Python.

لحساب التباين التجريبي لـ S و T بناءً على العينة الأصلية، يمكنك استخدام مكتبة numpy في لغة البرمجة بايثون

```python
import numpy as np


np.random.seed(123)
S_values = []
T_values = []

for _ in range(1000):

    sample = np.random.exponential(scale=1/2, size=20)
    S = 1 / np.mean(sample)
    T = 20 * np.min(sample)
    S_values.append(S)
    T_values.append(T)

S_values = np.array(S_values)
T_values = np.array(T_values)

var_S_empirical = np.var(S_values, ddof=1)
var_T_empirical = np.var(T_values, ddof=1)

print("Empirical variance of S:", var_S_empirical)
print("Empirical variance of T:", var_T_empirical)
```
