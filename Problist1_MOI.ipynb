```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123456)
data_pois = np.random.poisson(5, 100)

plt.plot(range(len(data_pois)), data_pois, 'bo')
plt.xlabel('Index')
plt.ylabel('Random Variable')
plt.title('Poisson Distribution Sample')
plt.show()
```

```python
import pandas as pd

frequency_table = pd.Series(data_pois).value_counts().sort_index()
print(frequency_table)
```

```python
import statsmodels.api as sm

ecdf = sm.distributions.ECDF(data_pois)
x = np.linspace(min(data_pois), max(data_pois))
y = ecdf(x)

plt.step(x, y)
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Empirical Cumulative Distribution Function')
plt.show()
```

```python
sample_mean = np.mean(data_pois)
poisson_prob = [np.exp(-sample_mean) * (sample_mean ** k) / np.math.factorial(k) for k in range(max(data_pois)+1)]

print("Sample Mean:", sample_mean)
print("Poisson Probability:", poisson_prob)
```

```python
plt.subplot(1, 2, 1)
frequency_table.plot(kind='bar')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Observed Values')

plt.subplot(1, 2, 2)
predicted_freq = [len(data_pois) * prob for prob in poisson_prob]
plt.bar(range(len(poisson_prob)), predicted_freq)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Values Predicted using Estimated Mean')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

data = pd.read_csv('faithful.csv')
print(data.head())
```

```python
plt.hist(data['waiting'], bins=30)
plt.xlabel('Waiting Time')
plt.ylabel('Frequency')
plt.title('Histogram of Waiting Time')
plt.show()
```

```python
plt.subplot(1, 3, 1)
plt.hist(data['waiting'], bins=40)
plt.xlabel('Waiting Time')
plt.ylabel('Frequency')
plt.title('40 Bins')

plt.subplot(1, 3, 2)
plt.hist(data['waiting'], bins=20)
plt.xlabel('Waiting Time')
plt.ylabel('Frequency')
plt.title('20 Bins')

plt.subplot(1, 3, 3)
plt.hist(data['waiting'], bins=5)
plt.xlabel('Waiting Time')
plt.ylabel('Frequency')
plt.title('5 Bins')

plt.tight_layout()
plt.show()

plt.scatter(data['waiting'], data['eruptions'])
plt.xlabel('Waiting Time')
plt.ylabel('Eruption Time')
plt.title('Scatter Plot of Eruption Time vs. Waiting Time')
plt.show()
```

```python
correlation = data['eruptions'].corr(data['waiting'])
print("Correlation Coefficient:", correlation)
```

```python
plt.boxplot(data['eruptions'])
plt.ylabel('Eruption Time')
plt.title('Box Plot of Eruption Time')
plt.show()
```

```python
data.boxplot('eruptions', by='waiting')
plt.xlabel('Waiting Time')
plt.ylabel('Eruption Time')
plt.title('Box Plot of Eruption Time Grouped by Waiting Time')
plt.suptitle('')
plt.show()
```

```python
covariance = data['eruptions'].cov(data['waiting'])
print("Covariance:", covariance)
```
